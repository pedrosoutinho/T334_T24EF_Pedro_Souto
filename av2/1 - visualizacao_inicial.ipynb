```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("dataset_19.csv")
```

```python
df.info()
```

```python
plt.figure(figsize=(10, 6))
plt.title("Distribuição do tipo de processador")
sns.countplot(data=df, x='tipo_processador')
plt.xlabel("Tipo de processador")
plt.ylabel("Frequência")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title("Distribuição do tipo de sistema operacional")
sns.countplot(data=df, x='sistema_operacional')
plt.xlabel("Tipo de sistema operacional")
plt.ylabel("Frequência")
plt.xticks(rotation=45)
plt.show()
```

```python
# plot the distribution of processors when system is MacOS

plt.figure(figsize=(10, 6))
plt.title("Distribuição do tipo de processador quando o sistema é MacOS")
sns.countplot(data=df[df['sistema_operacional'] == 'MacOS'], x='tipo_processador')
plt.xlabel("Tipo de processador")
plt.ylabel("Frequência")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title("Distribuição da latência em milissegundos")
sns.histplot(df['latencia_ms'].dropna(), bins=20)
plt.xlabel("Latência (ms)")
plt.ylabel("Frequência")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title("Distribuição do tempo de resposta em milissegundos")
plt.xlabel("Tempo de resposta (ms)")
sns.histplot(df['tempo_resposta'].dropna(), bins=20)
plt.ylabel("Frequência")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title("Relação entre latência e tempo de resposta")
sns.scatterplot(x=df['latencia_ms'], y=df['tempo_resposta'])
plt.xlabel("Latência (ms)")
plt.ylabel("Tempo de Resposta (ms)")
plt.show()
```

```python
df['latencia_ms'].corr(df['tempo_resposta'])
```

```python
plt.figure(figsize=(10, 6))
plt.title("Relação entre tipo de HD e tempo de resposta")
sns.boxplot(x=df['tipo_hd'], y=df['tempo_resposta'])
plt.xlabel("Tipo de HD")
plt.ylabel("Tempo de Resposta (ms)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title("Relação entre tipo de processador e tempo de resposta")
sns.boxplot(x=df['tipo_processador'], y=df['tempo_resposta'])
plt.xlabel("Tipo de Processador")
plt.ylabel("Tempo de Resposta (ms)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title("Relação entre armazenamento e tempo de resposta")
sns.boxplot(x=df['armazenamento_tb'], y=df['tempo_resposta'])
plt.xlabel("Armazenamento (TB)")
plt.ylabel("Tempo de Resposta (ms)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title("Distribuição de cpu cores")
sns.countplot(data=df, x='cpu_cores')
plt.xlabel("Cores de CPU")
plt.ylabel("Frequência")
plt.xticks(rotation=45)
plt.show()
```
